```python
import datetime,time
import pytz
import pandas as pd
from dateutil.relativedelta import *
```

```python
concerts = pd.read_csv("tz_list.csv")
```

```python
concerts
```

```python
for row in concerts.iterrows():
    czone = pytz.timezone(row[1]["IANA"])
    istzone = pytz.timezone('Asia/Kolkata')
    pdtzone = pytz.timezone('America/Costa_Rica')
    ctime = czone.localize(datetime.datetime.strptime(row[1]["ShowTime"],'%Y-%m-%d %H:%M:%S'))
    ctime_ist = ctime.astimezone(istzone)
    ctime_pdt = ctime.astimezone(pdtzone)
    isttime = datetime.datetime.now(istzone)
    pdttime = datetime.datetime.now(pdtzone)
    istdiff = relativedelta(ctime_ist, isttime)
    pdtdiff = relativedelta(ctime_pdt, pdttime)
#     istdiff = ctime_ist - isttime
#     pdtdiff = ctime_pdt - pdttime
    
    print(row[1]["City"])
    print("ShowTime in IST",ctime_ist)
    print("ShowTime in PDT",ctime_pdt)
    print("Local Time in IST ", isttime)
    print("Local Time in PDT ", pdttime)
    print("IST Diff ", istdiff)
    print("PDT Diff ", pdtdiff)
    print('=========================================================================================')
```
